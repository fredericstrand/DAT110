```python
import numpy as np
import matplotlib.pyplot as plt


g = 9.81 #m/s^2
mass = 20.0 #kg

def vertical_normal_force(mass):
    N = mass * g
    return N

normal_force_magnitude = vertical_normal_force(mass)
```

```python
'''
Exercise 1) define a function with two parameters that return the force done by friction with a coefficent mu
on a horizontal plane, based on the normal force
'''

def force_by_friction(normal_force_magnitude, mu):
    return mu * normal_force_magnitude
```

```python
'''
Exercise 2) create a function called kinetic_energy which return the kinetic energy. This function should have 
two parameters.
'''
def kinetic_energy(mass, velocity):
    return 0.5 * mass * velocity**2
```

```python
'''
Exercise 3) We know that a curling stone that has a speed of 1.0m/s travels exactly 6.4m and stops right at the
center of the curling target. We also know that the only force acting on the curling stone as it travels is the 
friction force. Find the friction coefficient mu.
'''
mu = kinetic_energy(mass,1.0) / (mass * g * 6.4)
mu
```

```python
'''
Exercise 4) Knowing the friction coefficient mu, plot the speed of the curling stone as it moves from the
starting line (hog line) starting with 1.0m/s, all the way to the centre of the target (at 6.4 metres).
Define a function for the speed of the curling stone. It should be a function called 'speed_curling_stone' of
four parameters: the starting kinetic energy, the coefficient, the mass, and the distance traveled.
'''

K = kinetic_energy(mass, 1.0)
L = 6.4

def speed_curling_stone(K, mass, mu, L):
    
    final_kinetic_energy = K - (force_by_friction(normal_force_magnitude, mu) * L)
    
    return np.sqrt(2*final_kinetic_energy / mass)

distance = np.linspace(0.0,6.4,100)
speed = speed_curling_stone(K, mass, mu, distance)

plt.plot(distance, speed)
plt.ylabel('Speed of the curling stone (m/s)')
plt.xlabel('Distance from the starting line(m)')
plt.title('Speed of the curling stone vs distance')
plt.grid(True)
plt.show();
    
```
