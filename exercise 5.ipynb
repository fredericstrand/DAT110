```python
import pandas as pd
import matplotlib.pyplot as plt
import os

os.makedirs('plots-exercise-5')

forest_df = pd.read_csv('data/tabell.csv',delimiter=';')
forest_df = forest_df.T
forest_df.columns = forest_df.iloc[0]
forest_df = forest_df.iloc[1:]
forest_df = forest_df.applymap(lambda x: float(x.replace(" ", "")) if isinstance(x, str) else x)
forest_df
```

```python
# Remove spaces and convert to float
properties_row = forest_df.iloc[1, 2:].apply(lambda x: float(x.replace(' ', '')) if isinstance(x, str) else x)

# Get the column names starting from the third column
columns = forest_df.columns[2:]

# Plotting Histogram
plt.figure(figsize=(10, 6))
plt.bar(columns, properties_row)
plt.title('Number of Properties Distribution')
plt.xlabel('Area Categories')
plt.ylabel('Number of Properties')

# Plotting Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(properties_row, labels=columns)
plt.title('Number of Properties Distribution')
plt.tight_layout()
```

```python
# Remove spaces and convert to float
forest_row = forest_df.iloc[3, 2:].apply(lambda x: float(x.replace(' ', '')) if isinstance(x, str) else x)

# Get the column names starting from the third column
columns = forest_df.columns[2:]

# Plotting Histogram
plt.figure(figsize=(10, 6))
plt.bar(columns, forest_row)
plt.title('Number of Productive Forest area size categories')
plt.xlabel('Forest Area Categories')
plt.ylabel('Number of Properties')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Plotting Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(forest_row, labels=columns, autopct='%1.1f%%', startangle=140)
plt.title('Number of Properties Distribution')
plt.axis('equal')
```
